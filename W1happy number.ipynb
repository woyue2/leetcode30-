```c
#include <stdio.h>
#include <stdbool.h>
int next_res(int num){
    int res = 0;
    while (num!= 0){
        int yushu = num % 10;
        num /= 10;
        res += yushu * yushu;
    }
    return res;
}

bool isHappy(int num){
    int slow = num;
    int fast = num;
    //  闭环点
    //  如果没有闭环点，就不会相遇；
    //  一旦相遇，就是有闭环点，这个时候刚好是超过一圈； -- 有点像时空电影的感觉
    do {
    slow = next_res(slow);
    fast = next_res(next_res(fast));
    } while (slow != fast);
    
    return slow == 1;
}

int main(){
    printf("%d", isHappy(19));
    return 0;
}
```

```c
#include <stdio.h>
#include <stdlib.h>  //结构体所需要

struct ListNode {
    int val;
    struct ListNode *next;
};

struct ListNode *detectCycle(struct ListNode *head) {
    // head slow fast都是指针
    struct ListNode *slow = head;
    struct ListNode *fast = head;

    // 如果只检查 fast 是否为 NULL，而不检查 fast->next 是否为 NULL，
    // 可能会导致在访问 fast->next 时发生空指针解引用错误。
    
    // check有没有环
    while (fast != NULL && fast->next != NULL) {
        slow = slow->next;
        fast = fast->next->next;
        if (slow == fast) {
            break;
        }
    } 
    // 这以后slow == fast

    // fast == NULL || fast->next == NULL
    if (fast == NULL || fast->next == NULL) {
        return NULL;
    } 
    
    // slow == fast
    slow == headb
    while (slow != fast) {
        slow = slow->next;
        fast = fast->next;
    }
    // printf("slow 的地址: %p\n", (void *)slow);
    return slow;
}

int main() {
    // 创建链表节点
    struct ListNode *head  = (struct ListNode *)malloc(sizeof(struct ListNode));
    struct ListNode *node1 = (struct ListNode *)malloc(sizeof(struct ListNode));
    struct ListNode *node2 = (struct ListNode *)malloc(sizeof(struct ListNode));
    struct ListNode *node3 = (struct ListNode *)malloc(sizeof(struct ListNode));
    struct ListNode *node4 = (struct ListNode *)malloc(sizeof(struct ListNode));

    // 构建链表
    head->val = 3;
    head->next = node1;
    node1->val = 2;
    node1->next = node2;
    node2->val = 0;
    node2->next = node3;
    node3->val = 4;
    node3->next = node4;
    node4->val = 5;
    node4->next = node2;  // 创建环，将node4的next指向node2

    // 检测环的起始节点
    struct ListNode *cycleStart = detectCycle(head);

    // 打印结果
    if (cycleStart != NULL) {
        printf("链表中存在环，环的起始节点值为: %d\n", cycleStart->val);
    } else {
        printf("链表中不存在环\n");
    }
    printf("head 的地址:  %p\n", (void *)head);
    printf("node1 的地址: %p\n", (void *)node1);
    printf("node2 的地址: %p\n", (void *)node2);
    printf("node2 的地址: %p\n", (void *)node3);
    printf("结构体大小: %d\n", sizeof(struct ListNode));

    

    // 释放内存
    free(head);
    free(node1);
    free(node2);
    free(node3);
    free(node4);
    
    
    printf("head 的地址:  %p\n", (void *)head);

    return 0;
}
```
